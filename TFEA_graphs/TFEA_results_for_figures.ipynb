# Read in TFEA results

```python
def readresultstable(filename):
    df = pd.read_csv(filename, sep="\t", index_col=False)
    return df
 
def abovecuttoff(row, sample):
    pval= row[pvalcolumn]
    if pval<=TFEAcutoffVal:
        col="Sig"
    else:
        col="N.S."
    return col
```

# Plot MA

```python

def MAplotresults_labeledpoints(filename, yshifts=[10,10,-10,0], xshifts=[10,5,-10,-20],outfilename="none", labelcolor="red"):
    samplename=filename.split("/")[-2]
    print (samplename)
    df = readresultstable(filename)
    df["cutoff"]= df.apply(lambda row : abovecuttoff(row, samplename), axis = 1)
    fig = px.scatter(df, y=df[escorecolumn], x=np.log10(df['Events']), hover_data=['#TF', 'P-adj', 'Corrected P-adj'], color=df["cutoff"],color_discrete_sequence=[labelcolor, "#D3D3D3"])
    fig.update_layout({'plot_bgcolor': 'rgba(0, 0, 0, 0)','paper_bgcolor': 'rgba(0, 0, 0, 0)',})
    minidf = df[df["#TF"].isin(marktheseTFs.keys())]
    print(df.head())
    minidf = minidf[minidf["Corrected P-adj"]<=TFEAcutoffVal]
    for i, idxval in enumerate(minidf.index):
        yshift = yshifts[i]
        xshift = xshifts[i]
        print (df.loc[idxval,"#TF"])
        fig.add_annotation(x=np.log10(df.loc[idxval,"Events"]), y=df.loc[idxval,escorecolumn], text=marktheseTFs[df.loc[idxval,"#TF"]],showarrow=False,yshift=yshift, xshift=xshift, font=dict(
        family="sans serif",
        size=30,
        color=labelcolor
    ))
    fig.update_layout(xaxis_title="mean signal for this TF",yaxis_title="TF Enrichment Score",
    font=dict(
        family="sans serif",
        size=30))
    fig.update_layout(showlegend=False)
    if outfilename!="none":
        fig.write_image(outfilename,engine='kaleido')
    fig.show()
    return df
```

# Plot scatter 

```python
def checkcat(row, sample1, sample2):
    pval1= row[pvalcolumn+"_"+sample1]
    pval2= row[pvalcolumn+"_"+sample2]
    if pval1<=TFEAcutoffVal and pval2<=TFEAcutoffVal:
        cat = "adj-pval<="+str(TFEAcutoffVal)+", both comparisons"
    elif pval1<=TFEAcutoffVal:
        cat = "adj-pval<="+str(TFEAcutoffVal)+", "+sample1
    elif pval2<=TFEAcutoffVal:
        cat = "adj-pval<="+str(TFEAcutoffVal)+", "+sample2
    else:
        cat="N.S."
    return cat


def mergedfs(filename1, filename2, label1, label2):
    df1 = readresultstable(filename1)
    df2 = readresultstable(filename2)
    df = df1.merge(df2, left_on='#TF', right_on='#TF',suffixes=('_'+label1, '_'+label2))
    df["category"] =  df.apply(lambda row : checkcat(row, label1, label2), axis = 1)
    return df, label1, label2

def merge_multi_dfs(samplesdict):
    dfs = []
    for i, sample in enumerate(sorted(samplesdict.keys())):
        df = readresultstable(sample)
        samplename = samplesdict[sample]
        df.columns = [samplename+"_"+col if col!="#TF" else col for col in df.columns]
        if i==0:
            masterdf = df
        else:
            masterdf = masterdf.merge(df, left_on='#TF', right_on='#TF')
    return masterdf




def scattertwo(filename1, filename2, label1, label2, columname=escorecolumn, yshifts=[0,0,0,0], xshifts=[0,0,0,0],outfilename="none"):
    df, sample1, sample2 = mergedfs(filename1, filename2, label1, label2)
    xdata = columname+"_"+sample1
    ydata = columname+"_"+sample2
    df = df.sort_values(by="category")
    colordict = {"adj-pval<="+str(TFEAcutoffVal)+", both comparisons":"rgb(255,0,0)", "adj-pval<="+str(TFEAcutoffVal)+", "+label1:samplecolor[label1], "adj-pval<="+str(TFEAcutoffVal)+", "+label2:samplecolor[label2], "N.S.":"#D3D3D3"}
    fig = px.scatter(df, x=df[xdata], y=df[ydata], color="category", hover_data=['#TF'], color_discrete_map=colordict)
    fig.update_layout(shapes = [{'type': 'line', 'y0': -0.4, 'y1': 0.4, 'x0': -0.4, 'x1': 0.4,"line":{"color":"rgb(200, 200, 200)"}}])
    fig.update_layout({'plot_bgcolor': 'rgba(0, 0, 0, 0)','paper_bgcolor': 'rgba(0, 0, 0, 0)',})
    fig.update_layout(showlegend=False)
    minidf = df[df["#TF"].isin(marktheseTFs.keys())]
    for i, idxval in enumerate(minidf.index):
        yshift = yshifts[i]
        xshift = xshifts[i]
        colorlookup = df.loc[idxval,"category"]
        colorlookup = colordict[colorlookup]
        print (df.loc[idxval,"#TF"])
        fig.add_annotation(x=df.loc[idxval,xdata], y=df.loc[idxval,ydata], text=marktheseTFs[df.loc[idxval,"#TF"]],showarrow=False,yshift=yshift, xshift=xshift, font=dict(
         family="sans serif",
         size=30,
         color=colorlookup))
    fig.update_layout(xaxis_title="Disomic HS induced changes",yaxis_title="Trisomic HS induced changes",
    font=dict(
        family="sans serif",
        size=30))
    if outfilename!="none":
        fig.write_image(outfilename,engine='kaleido')
    fig.show()
    return df
```

# Plot pvalues all 4 comparisons

```python
HScompares = {ATACErichs:"ATAC-seq D21 Heatshock", ATACEthanhs: "ATAC-seq T21 Heatshock", PROErichs:"PRO-seq D21 Heatshock", PROEthanhs:"PRO-seq T21 Heatshock"}             


def createadf_multisample_onecol(column=pvalcolumn, sampledict=HScompares):
    a = merge_multi_dfs(sampledict)
    a.index = a["#TF"]
    colsneeded = [col for col in a.columns if col.endswith("_"+column)]    
    return a[colsneeded]

def createmergedataframeforplotHSpvals():
    sampledict=HScompares
    a= createadf_multisample_onecol(column=pvalcolumn, sampledict=sampledict)
    a = a[a.min(axis=1)<=TFEAcutoffVal] #filter only TF's above the cuttof
    no_zero = a.min(axis=1)!=0
    m = a[no_zero].min().min()
    a = a.replace(to_replace=0,value=m)
    a = a.apply(transformpvals)
    dfmerge =a
    id_vars = ["#TF"]
    value_vars = dfmerge.columns.tolist()
    dfmerge["#TF"] = dfmerge.index
    dfmergelong = pd.melt(dfmerge, id_vars=id_vars, value_vars=value_vars)
    dfmergelong[["sample", "value_type"]] = dfmergelong["variable"].str.split("_", expand=True)
    dfmergelong = dfmergelong[["#TF", "sample", "value"]]
    dfmergelong.columns = ["#TF", "sample", pvalcolumn]
    b = createadf_multisample_onecol(column=escorecolumn, sampledict=sampledict)
    b = b.reindex(dfmerge.index.tolist())
    value_vars = b.columns.tolist()
    b["#TF"] = b.index
    blong = pd.melt(b, id_vars=id_vars, value_vars=value_vars)
    blong[["sample", "value_type"]] = blong["variable"].str.split("_", expand=True)
    blong = blong[["#TF", "sample", "value"]]
    blong.columns = ["#TF", "sample", escorecolumn]
    abmerge = dfmergelong.merge(blong, left_on=["#TF", "sample"], right_on=["#TF", "sample"])
    abmerge["sig"] = abmerge[pvalcolumn]<=transformpvals(TFEAcutoffVal)
    abmerge["sig"] = abmerge["sig"].astype(int)
    grouped_multiple = abmerge.groupby(['#TF']).agg({pvalcolumn: ['mean', 'min', 'max',], "sig":['sum'], escorecolumn: ['mean', 'min', 'max', "sum"]})
    grouped_multiple.columns = [pvalcolumn+'_mean', pvalcolumn+'_min', pvalcolumn+'_max', "sig_count", escorecolumn+"_mean", escorecolumn+"_min", escorecolumn+"_max", escorecolumn+"_sum"]
    grouped_multiple = grouped_multiple.reset_index()
    grouped_multiple = grouped_multiple.sort_values(["sig_count", escorecolumn+'_sum'], ascending=[True, False])
    sorter = grouped_multiple["#TF"].tolist()
    sorterIndex = dict(zip(sorter, range(len(sorter))))
    abmerge['TF_Rank'] = abmerge['#TF'].map(sorterIndex)
    abmerge = abmerge.sort_values(["TF_Rank"], ascending=[False])
    return abmerge



def sizetrans(row):
    s = row[pvalcolumn]
    s2 = 1+(s*10)
    return s2

def HSpvaluesTFEAforpaper(outfilename="none"):
    abmerge = createmergedataframeforplotHSpvals()
    abmerge[["justTF", "org"]]=abmerge["#TF"].str.split("_", n=1, expand=True)
    abmerge[pvalcolumn+"_transform"] = abmerge.apply(lambda row : sizetrans(row), axis = 1)
    fig = px.scatter(abmerge, x="sample", y="justTF",color=escorecolumn,size=pvalcolumn+"_transform",color_continuous_scale=px.colors.diverging.balance,
                                 color_continuous_midpoint=0, symbol="sig", category_orders={"sample": ['ATAC-seq D21 Heatshock', 'ATAC-seq T21 Heatshock', 'PRO-seq D21 Heatshock', 'PRO-seq T21 Heatshock']})
    fig.update_layout(height=1000, width=500, plot_bgcolor='rgba(0,0,0,0)')
    fig.update_xaxes(showgrid=False, zeroline=False)
    fig.update_yaxes(showgrid=False, zeroline=False)
    fig.update_coloraxes(colorbar_len=.7)
    fig.update_layout(font=dict(
        size=18))
    fig.show()
    if outfilename!="none":
        fig.write_image(outfilename,engine='kaleido')

def transformpvals(pval):
    transpval = (-1*np.log10(pval))**(1/6)
    return transpval

def revtransformpvals(transpval):
    pval = (10**((transpval**6)/-1))
    return pval
```
